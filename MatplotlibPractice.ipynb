```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# to execute and see results inline
%matplotlib inline
```

```python
#create input arrays that will be used to map
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# functional method to create plot
plt.plot(x,y)
#if not on jupyter, enter plt.slow()
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(x,y)
```

```python
#object oriented approach
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('larger plot')
axes2.set_title('smaller plot')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#linewidth takes the widthness factor
ax.plot(x,y,color='green',linewidth=1,linestyle='--', marker = '+', markersize='10')
#possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1'
```
